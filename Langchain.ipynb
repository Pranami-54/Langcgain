```python
from langchain.document_loaders import WebBaseLoader

url = "https://brainlox.com/courses/category/technical"
loader = WebBaseLoader(url)
documents = loader.load()
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings.openai import OpenAIEmbeddings

vectorstore = Chroma.from_documents(
    documents,
    embedding=OpenAIEmbeddings(),
)
```

Create a FLASK RESTFUL API to handle the conversation

```python
from flask import Flask, request, jsonify
from langchain.chains import RetrievalQA
from langchain.llms import OpenAI # Import the OpenAI class from langchain


app = Flask(__name__)

# Set up the QA chain using the vector store
qa_chain = RetrievalQA.from_chain_type(llm=OpenAI(), chain_type="stuff", retriever=vectorstore.as_retriever())

@app.route('/chat', methods=['POST'])
def chat():
    user_input = request.json.get("message")
    if user_input:
        response = qa_chain.run(user_input)
        return jsonify({"response": response})
    return jsonify({"error": "No input provided"})

if __name__ == "__main__":
    app.run(debug=True)
```
